```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
X,y = make_blobs(n_samples = 500, n_features = 2, centers=5,random_state=3)
```

```python
print(X)
```

```python
print(y) #consists of numbers from 0-4 indicating which point belongs to which cluster
```

```python
fig = plt.figure(0)
plt.grid(True)
plt.scatter(X[:,0],X[:,1])
plt.show()
```

```python
fig = plt.figure(0)
plt.grid(True)
plt.scatter(X[:,0],X[:,1], c=y) #showing color
plt.show()
```

# K-Means Clustering

```python
k = 5
colors = ["green","red","blue","yellow","cyan"]
```

```python
#now we need to generate random clusters from -10 to 10
#np.random.random(X.shape[1])  this code will return points ranging from 0-1
10*(2*np.random.random((X.shape[1],))-1) #this will generate points ranging from -10 to 10
```

```python
clusters = {}
for i in range(k):
    center = 10*(2*np.random.random((X.shape[1],))-1)
    points = []
    cluster = {
        'center' : center,
        'points' : [],
        'color' : colors[i]
    }
    clusters[i] = cluster
```
