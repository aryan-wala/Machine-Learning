# K-Means Clustering

```python
k = 5
colors = ["green","red","blue","yellow","cyan"]
```

```python
#now we need to generate random clusters from -10 to 10
#np.random.random(X.shape[1])  this code will return points ranging from 0-1
10*(2*np.random.random((X.shape[1],))-1) #this will generate points ranging from -10 to 10
```

```python
clusters = {}
for i in range(k):
    center = 10*(2*np.random.random((X.shape[1],))-1)
    points = []
    cluster = {
        'center' : center,
        'points' : [],
        'color' : colors[i]
    }
    clusters[i] = cluster
```

```python
clusters
```

```python
plt.scatter(X[:,0],X[:,1],c = y)

for i in clusters:
    center = clusters[i]['center']
    plt.scatter(center[0],center[1], marker = "x", c = "magenta")
plt.show()
```

```python
# distance function
def distance(v1,v2):
    return np.sqrt(np.sum((v1-v2)**2))

v1 = np.array([1,2])
v2 = np.array([3,4])
print(distance(v1,v2))
```

```python
#implementing the E step
def assign_cluster():
    for i in range(X.shape[0]):
        dist = []
        
        curr_x = X[i]
        
        for j in range(k):
            dis = distance(curr_x,clusters[j]['center'])
            dist.append(dis)
        curr_cluster = np.argmin(dist) #argmin gives the index not the value
        clusters[curr_cluster]['points'].append(curr_x)
```

```python
assign_cluster()
```

```python
clusters
```

```python
def plot_cluster():
    for i in clusters:
        pts = np.array(clusters[i]['points'])
        try:
            plt.scatter(pts[:,0],pts[:,1],c = clusters[i]["color"])
        except:
            pass
        center = clusters[i]['center']
        plt.scatter(center[0],center[1],c = "magenta")
```

```python
plot_cluster()
```

```python
#implementing M step
def update_center():
    for i in range(k):
        pts = np.array(clusters[i]['points'])
        if(pts.shape[0] > 0):
            new_center = pts.mean(axis = 0) #axis =0 represnets 
            clusters[i]["center"] = new_center
            clusters[i]['points'] = []
```

```python

assign_cluster()
plot_cluster()
update_center()
```
