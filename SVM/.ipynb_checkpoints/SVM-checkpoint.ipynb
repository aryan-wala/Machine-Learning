```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("insurance.csv")
```

```python
df.head()
```

## Data cleaning

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.describe()    
```

## Feature Engineering

```python
Male = pd.get_dummies(df["sex"], drop_first=True, dtype=int) ## drop first willl drop the first column
```

```python
Male.info()
```

```python
Male
```

```python
df["sex"] = Male
```

```python
df.head()
```

```python
#Similarly for smoker
Smoker = pd.get_dummies(df["smoker"], drop_first=True, dtype=int)
df["smoker"] = Smoker
df.head()
```

```python
#For region
df["region"].unique()
```

```python
regions = pd.get_dummies(df["region"],dtype = int)
regions
df = pd.concat([df,regions], axis=1)
df.drop("region",axis = 1)
```
